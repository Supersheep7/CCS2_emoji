```python
df = df.drop([0, 19, 20])
df = df.drop([0, 1, 2, 3, 4, 5], axis=1)
```

```python
df_only_valence = df[['VALENCE']]
df_only_arousal = df[['AROUSAL']]
df_only_similarity = df[['SIMILARITY']]
df_only_relatedness = df[['RELATEDNESS']]
df_only_relatedness
```

```python
df_only_relatedness = df_only_relatedness.applymap(int)
df_only_similarity = df_only_similarity.applymap(int)
df_only_valence = df_only_valence.applymap(int)
df_only_arousal = df_only_arousal.applymap(int)

relatedness_means = df_only_relatedness.mean().tolist()
similarity_means = df_only_similarity.mean().tolist()
valence_means = df_only_valence.mean().tolist()
arousal_means = df_only_arousal.mean().tolist()
```

```python
arr1 = df_only_valence.applymap(int).values 
arr2 = df_only_arousal.applymap(int).values 
array_avg = (arr1 + arr2)/2

df_valence_arousal = pd.DataFrame(array_avg)
valence_arousal_means = df_valence_arousal.mean().tolist()
```

```python
arr1 = df_only_similarity.applymap(int).values 
arr2 = df_only_relatedness.applymap(int).values 
array_avg = (arr1 + arr2)/2

df_similarity_relatedness = pd.DataFrame(array_avg)
similarity_relatedness_means = df_similarity_relatedness.mean().tolist()
```

### Hypothesis testing

```python
from scipy.stats import t

ALPHA = 0.05
SAMPLE_SIZE = 17
DOF = SAMPLE_SIZE - 2

t_value = (average_correlation_arousal * np.sqrt(DOF))/np.sqrt(1 - average_correlation_arousal**2)
p = 1 - t.cdf(t_value, DOF)
```

### Mean std is around 0.84 (directly extracted on the xls)

```python
from scipy.stats import t

# Sample correlation coefficient
r = average_correlation_relatedness

# Fisher transformation
z = (np.log((1+r)/(1-r)))/2

# Sample size
N = 17

# Standard error of z
SE_z = 1 / (N - 3)**0.5

# Test statistic
test_statistic = z / SE_z

# Degrees of freedom
df = N - 3

print(test_statistic)

# Critical value for a two-tailed test at alpha = 0.05
alpha = 0.05
critical_value = t.ppf(1 - alpha, df)
print(critical_value)

# Test for significance
if abs(test_statistic) > critical_value:
    print("Correlation is statistically significant.")
else:
    print("Correlation is not statistically significant.")
```
